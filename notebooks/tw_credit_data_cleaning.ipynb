## Data cleaning

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/taiwan_credit.csv")
print(df.head())
print(df.info())
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.head()
print(df.isna().sum())
```

```python
df.columns = df.iloc[0]
df = df.iloc[1:].copy()
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
df = df.rename(columns={"default payment next month": "DEFAULT"})
df = df.apply(pd.to_numeric)
```

```python
df.describe()
```

```python
df["PAY_0"].value_counts() # there is -2 value which its meaning is not documented
```

```python
# finding out if -2 in PAY_0 has any special meaning by checking BILL_AMT2 statistics
print(df.loc[df["PAY_2"] == -2, "BILL_AMT2"].median())
print(df.loc[df["PAY_2"] == -2, "BILL_AMT2"].quantile([0.75, 0.9, 0.95, 0.99]))
# median of BILL_AMT2 for PAY_2 = -2 is 0, however the 90th percentile is quite high at 10830.10
```

```python
print(df.groupby("PAY_2")["BILL_AMT2"].median().sort_index())
print(df.groupby("PAY_2")["BILL_AMT2"].quantile(0.9).sort_index())

```

```python
df.groupby("PAY_2")["DEFAULT"].mean().sort_index() # -2 seems to have higher default rate than -1 so we keep it as is
```

##### Repayment status −2 corresponds to a distinct category whose precise semantics are not specified in the original documentation. Exploratory analysis shows that its default rates are comparable to other non-delinquent repayment states (−1, 0, 1) and clearly separated from delinquent categories. It is therefore retained as a valid non-delinquent in the analysis

```python
df["EDUCATION"].value_counts()
```

```python
# in the original documentation, EDUCATION values 0, 5, 6 are not defined
# we will group them into "others" category with value 4
df["EDUCATION"] = df["EDUCATION"].replace([0, 5, 6], 4)
df["EDUCATION"].value_counts().sort_index()
```

```python
df["MARRIAGE"].value_counts()
```

```python
# in the original documentation, MARRIAGE value 0 is not defined
# we will group it into "others" category with value 3
df["MARRIAGE"] = df["MARRIAGE"].replace(0, 3)
df["MARRIAGE"].value_counts().sort_index()
```

```python
df["SEX"].value_counts()
```

```python
df[["SEX", "EDUCATION", "MARRIAGE"]].describe()
```

##### Invalid categorical codes in EDUCATION (0, 5, 6) and MARRIAGE (0) were grouped into an “Others” category following common practice in prior analyses. This preserves all observations while avoiding undocumented categories

```python
df.isna().sum()
```

```python
df["DEFAULT"].value_counts(normalize=True)
```

```python
amount_cols = [c for c in df.columns if "AMT" in c]
df[amount_cols].describe().T[["min","max"]]
```

##### Statement balance variables (BILL_AMT1–BILL_AMT6) can take negative values, reflecting overpayments, refunds, or account credits rather than data errors.

```python
status_cols = [c for c in df.columns if c.startswith("PAY_") and "AMT" not in c]
{col: sorted(df[col].unique()) for col in status_cols}

```

At this stage, the dataset has been fully cleaned and validated for exploratory data analysis and baseline modeling. All variables are correctly typed, undocumented categories have been addressed, and no observations were removed. Further preprocessing steps will be applied as required by specific modeling approaches.
